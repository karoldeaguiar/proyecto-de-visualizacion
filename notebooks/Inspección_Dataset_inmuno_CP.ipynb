## Definición conceptual del índice de relevancia temática

Se construyó un **thematic relevance score** para estimar la **relevancia temática específica** de cada artículo respecto al núcleo conceptual del proyecto: *inmunoterapia perioperatoria y cirugía en cáncer de pulmón no microcítico resecable*.

Este índice es un **indicador derivado**, inspirado en técnicas clásicas de ponderación de términos como **TF-IDF**, y adaptado mediante **criterio experto** para su aplicación en un dominio clínico específico. Su objetivo es puramente exploratorio y está orientado a apoyar el análisis visual y la identificación de literatura central dentro del conjunto analizado.

En este contexto, el índice de relevancia temática mide:

- **centralidad semántica**, no calidad científica  
- **alineación temática**, no impacto clínico  
- **grado de focalización**, no nivel de evidencia  

El índice **no pretende evaluar la calidad metodológica ni el valor clínico** de los artículos, sino identificar aquellos más directamente relacionados con el foco temático del proyecto.

El cálculo del índice se basa en el análisis de **títulos y resúmenes**, identificando la presencia y el peso relativo de términos asociados a cuatro dimensiones clínicas clave:

1. Inmunoterapia  
2. Cirugía  
3. Contexto perioperatorio  
4. Cáncer de pulmón no microcítico resecable  

La combinación ponderada de estas dimensiones permite estimar la **centralidad temática** de cada artículo dentro del corpus analizado.

```python
import re

# Partimos del df completo
work = df.copy()

# Filtrar a registros con abstract disponible
work["abstract_summary"] = work["abstract_summary"].fillna("")
work["title"] = work["title"].fillna("")

work["has_abstract"] = work["abstract_summary"].str.strip().ne("")
work = work[work["has_abstract"]].copy()

# Texto base: título + abstract
work["text_base"] = (work["title"].str.strip() + " " + work[
    "abstract_summary"].str.strip()).str.strip()

# Diccionarios de términos (regex)
TERMS = {
    "surgery": [
        r"\bsurger(y|ies|ical)\b", r"\bresection\b", r"\bresectable\b",
        r"\blobectom(y|ies)\b", r"\bpneumonectom(y|ies)\b",
        r"\bthoracic\b", r"\boperative\b", r"\bvideo[- ]assisted\b", r"\bvats\b",
        r"\br0\b"
    ],
    "io": [
        r"\bimmunotherap(y|ies)\b", r"\bimmune checkpoint\b",
        r"\bcheckpoint inhibitor(s)?\b",
        r"\bpd-?1\b", r"\bpd-?l1\b", r"\bctla-?4\b",
        r"\bnivolumab\b", r"\bpembrolizumab\b", r"\bdurvalumab\b",
        r"\batezolizumab\b",
        r"\bsintilimab\b", r"\btoripalimab\b", r"\bipilimumab\b",
        r"\bcemiplimab\b"
    ],
    "periop": [
        r"\bperioperative\b", r"\bneoadjuvant\b", r"\badjuvant\b",
        r"\bpreoperative\b", r"\bpostoperative\b", r"\binduction\b"
    ],
    "nsclc": [
        r"\bnsclc\b", r"\bnon[- ]small cell lung cancer\b", r"\blung cancer\b",
        r"\bearly[- ]stage\b", r"\bstage (i|ii|iii|iv|1|2|3|4)\b",
        r"\bresectable\b"
    ],
}

COMPILED = {k: [re.compile(pat, flags=re.IGNORECASE) for pat in pats] for k,
            pats in TERMS.items()}


def count_matches(text: str, compiled_patterns) -> int:
    """
    Cuenta el número total de coincidencias de patrones en un texto.
    """
    total = 0
    for pat in compiled_patterns:
        total += len(pat.findall(text))
    return total


# Conteos por dimensión
for dim, pats in COMPILED.items():
    work[f"cnt_{dim}"] = work["text_base"].apply(
        lambda t: count_matches(t, pats))

# Normalización por longitud
work["n_tokens"] = work["text_base"].str.findall(r"\w+").apply(len).clip(lower=1)


def tf_score(counts: pd.Series, n_tokens: pd.Series) -> pd.Series:
    """
    Calcula una puntuación TF normalizada a partir de densidades de términos.
    """
    density = counts / n_tokens
    raw = np.log1p(density * 1000.0)
    mn, mx = raw.min(), raw.max()
    if mx == mn:
        return pd.Series(np.zeros(len(raw)), index=raw.index)
    return (raw - mn) / (mx - mn)


for dim in ["surgery", "io", "periop", "nsclc"]:
    work[f"score_{dim}"] = tf_score(work[f"cnt_{dim}"], work["n_tokens"])

# Ponderaciones (las mantengo como las tienes)
weights = {"periop": 0.40, "io": 0.30, "surgery": 0.20, "nsclc": 0.10}

# Score continuo + normalizado [0,1]
work["thematic_relevance_raw"] = (
    weights["surgery"] * work["score_surgery"] +
    weights["io"]      * work["score_io"] +
    weights["periop"]  * work["score_periop"] +
    weights["nsclc"]   * work["score_nsclc"]
)

mn, mx = work["thematic_relevance_raw"].min(), work[
    "thematic_relevance_raw"].max()
work["thematic_relevance_score"] = 0.0 if mx == mn else (work[
    "thematic_relevance_raw"] - mn) / (mx - mn)

# Top 10
cols_show = ["source", "id", "year", "title", "url", "thematic_relevance_score",
             "cnt_surgery", "cnt_io", "cnt_periop", "cnt_nsclc"]

top10 = work.sort_values("thematic_relevance_score", ascending=False)[cols_show].head(10)
top10
```

```python
# Exportar tabla para Tableau
# Unimos el TIF al dataset completo
df_out = df.merge(
    work[["id", "thematic_relevance_score"]],
    on="id",
    how="left"
)

# Se filtra el año 2026, que está comenzando
df_out = df_out[df_out["year"] < 2026].copy()

# Limpieza del texto paises

# Lista de países válidos que aparecen en tu dataset
VALID_COUNTRIES = [
    "China", "United States", "Japan", "Italy", "Germany", "Spain", "France",
    "Canada", "United Kingdom", "Switzerland", "South Korea", "Australia",
    "The Netherlands", "Austria", "Belgium", "Turkey", "Egypt", "Ireland",
    "Malaysia", "Romania", "Jordan", "Denmark", "Norway", "Brazil", "Argentina",
    "Iran", "Singapore", "Thailand", "India", "Israel", "Portugal", "Poland",
    "Sweden", "Hungary", "Greece", "Lithuania", "Mexico", "New Zealand",
    "Czech Republic", "Taiwan", "Austria", "Belgium"
]


def clean_country_advanced(value):
    """
    Normaliza y clasifica el país, detectando estudios multicéntricos.
    """
    if pd.isna(value):
        return None

    text = str(value)

    # Caso multicéntrico (coma entre países)
    if "," in text:
        return "Multicenter"

    # Normalizar texto
    text = text.replace("People'S Republic Of China", "China")
    text = text.replace("People's Republic of China", "China")
    text = text.replace("PR China", "China")
    text = text.replace("Deutschland", "Germany")
    text = text.replace("Turkiye", "Turkey")

    # Quedarse con lo anterior al primer punto
    text = text.split(".")[0].strip()

    # Buscar país válido
    for country in VALID_COUNTRIES:
        if country.lower() in text.lower():
            return country

    return None


df_out["countries"] = df_out["countries"].apply(clean_country_advanced)

# Limpieza de los títulos
df_out["title"] = df_out["title"].str.replace(r"[\[\]]", "", regex=True)
```

```python
# Guardar
df_out.to_excel("dataset_for_tableau.xlsx", index=False)

print("OK -> dataset_for_tableau.xlsx")
```
